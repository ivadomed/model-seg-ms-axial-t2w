```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure and subplots
fig, axs = plt.subplots(2, 1, figsize=(20, 15))

# First subplot (train set lesion volume distribution)
ax = axs[0]

# Filter values greater than 800
df_tum_train_filtered = df_tum_train[df_tum_train["Object volume"] <= 800]
df_nyu_train_filtered = df_nyu_train[df_nyu_train["Object volume"] <= 800]

df_tum_train_filtered["Object volume"].hist(ax=ax, alpha=0.5, label='tum', bins=100)
df_nyu_train_filtered["Object volume"].hist(ax=ax, alpha=0.5, label='nyu', bins=100)

# Adding titles and labels using set_title and set_ylabel
ax.set_title("Train Set Lesion Volume Distribution", fontsize=18)
ax.set_xlabel("Lesion Volume [mm³]", fontsize=LABEL_FONT_SIZE)
ax.set_ylabel("Lesion Frequency", fontsize=LABEL_FONT_SIZE)

# Set ticks and limits
ax.tick_params(axis='x', labelsize=TICK_FONT_SIZE)
ax.tick_params(axis='y', labelsize=TICK_FONT_SIZE)
ax.set_xlim(left=0)

# Add legend
ax.legend()

# Second subplot (test set lesion volume distribution)
ax = axs[1]
df_tum_testset_filtered = df_tum_testset[df_tum_testset["Object volume"] <= 800]
df_bwh_testset_filtered = df_bwh_testset[df_bwh_testset["Object volume"] <= 800]
df_uscf_testset_filtered = df_uscf_testset[df_uscf_testset["Object volume"] <= 800]

# Plot the filtered data
df_tum_testset_filtered["Object volume"].hist(ax=ax, alpha=0.5, label='tum', bins=100)
df_bwh_testset_filtered["Object volume"].hist(ax=ax, alpha=0.5, label='bwh', bins=100)
df_uscf_testset_filtered["Object volume"].hist(ax=ax, alpha=0.5, label='ucsf', bins=100)

# Adding titles and labels using set_title and set_ylabel
ax.set_title("Test Set Lesion Volume Distribution", fontsize=18)
ax.set_xlabel("Lesion Volume [mm³]", fontsize=LABEL_FONT_SIZE)
ax.set_ylabel("Lesion Frequency", fontsize=LABEL_FONT_SIZE)

# Set ticks and limits
ax.tick_params(axis='x', labelsize=TICK_FONT_SIZE)
ax.tick_params(axis='y', labelsize=TICK_FONT_SIZE)
ax.set_xlim(left=0)

# Add legend
ax.legend()

# Save the figure
plt.savefig("distribution_smaller_equal_800.png", dpi=300, bbox_inches="tight")
```
